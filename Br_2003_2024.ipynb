# VITÓRIAS POR TIME

```python
## mostrar o total de vitorias por time
vitorias = df[df['vencedor'] != 'empate']['vencedor'].value_counts()
display (vitorias)
```

```python
plt.figure(figsize=(25,10))
vitorias.plot(kind='bar', color='blue')

plt.title('Número de Vitórias por Time', fontsize= 22)
plt.xlabel('Time', fontsize= 22)
plt.ylabel('Vitórias', fontsize= 22)
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=25)


plt.tight_layout()
plt.show()
```

# GOLS MARCADOS

```python
# mostrar gols marcados como mandante
gols_mandante = df.groupby("mandante")["mandante_Placar"].sum()
display(gols_mandante)
```

```python
# mostrar gols marcados como visitante
gols_visitante = df.groupby("visitante")["visitante_Placar"].sum()
display(gols_visitante)
```

```python
# mostrar total de gols marcados entre 2003 e 2024 no brasileirao em ordem
total_gols = gols_mandante.add(gols_visitante, fill_value=0)
df_gols_marcados = total_gols.reset_index()
df_gols_marcados.columns = ['Times', 'Gols Marcados']
df_gols_marcados = df_gols_marcados.sort_values(by='Gols Marcados', ascending=False)
display(df_gols_marcados)
```

```python
## mostrar em barras o total de gols marcados por times que jogavam no brasileirao serie A entre 2003 - 2024

plt.figure(figsize=(25,10))
bars = plt.bar(df_gols_marcados['Times'], df_gols_marcados['Gols Marcados'], color='blue')
plt.title('Total de Gols Marcados por Time', fontsize = 22)
plt.xlabel('Times', fontsize = 22)
plt.ylabel('Gols Marcados', fontsize = 22)
plt.xticks(rotation=45, ha='right', fontsize=14)

## colocar os valores em cima das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 7,
             f'{int(height)}',
             ha='center',
             fontsize=11)
plt.tight_layout()
plt.show()
```

# GOLS SOFRIDOS

```python
## mostrar gols sofridos como mandante
gols_sofridos_mandante = df.groupby("mandante")["visitante_Placar"].sum()
display(gols_sofridos_mandante)
```

```python
## mostrar gols sofridos como visitante
gols_sofridos_visitante = df.groupby("visitante")["mandante_Placar"].sum()
display(gols_sofridos_visitante)
```

```python
total_gols_sofridos = gols_sofridos_mandante.add(gols_sofridos_visitante, fill_value=0)
df_gols_sofridos = total_gols_sofridos.reset_index()
df_gols_sofridos.columns = ['Times', 'Gols Sofridos']
df_gols_sofridos = df_gols_sofridos.sort_values(by='Gols Sofridos', ascending=False)
display(df_gols_sofridos)
```

```python
plt.figure(figsize=(25,10))
bars = plt.bar(df_gols_sofridos['Times'], df_gols_sofridos['Gols Sofridos'], color='red')
plt.title('Total de Gols Sofridos por Time', fontsize = 22)
plt.xlabel('Times', fontsize = 22)
plt.ylabel('Gols Sofridos', fontsize = 22)
plt.xticks(rotation=45, ha='right', fontsize=14)

# Colocar os valores em cima das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 7,
             f'{int(height)}',
             ha='center',
             fontsize=11)
plt.tight_layout()
plt.show()
```

# ARENAS

```python
## criar grafico pizza com a % de vezes que foram jogadas em cada estadio
contagem_arenas = df['arena'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(
    contagem_arenas,
    labels=contagem_arenas.index,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 10}
)
plt.title('Top 10 Arenas', fontsize=12)
plt.axis('equal')
plt.show()
```
